Im nachfolgenden Schritt wird eine einfache Funktion in R erstellt. Hierbei wird über eine Zuweisung eine neue Funktion temporär definiert. Die Funktion kann absofort über plusZwei() abgerufen werden. Die Funktionsparameter stehen hinter "function()" in den runden Klammern. Bei unserer ersten einfachen Funktion wollen wir lediglich die Zahl zwei auf ein Objekt / Zahl aufaddieren.

```r
plusZwei <- function(x, ...) {
    x <- x + 2
    return(x)
}
```

Nun ist die Funktion temporär aktiviert und wir können diese nun in R verwenden.

```r
plusZwei(a); plusZwei(6)
```
